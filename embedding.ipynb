```python
from __future__ import absolute_import
from __future__ import division
from __future__ import print_function

import pickle5 as pickle
import numpy as np

DATA_DIR = '../data/'
```

```python
# Load pre-trained embeddings
embeddings = {}
with open(f'{DATA_DIR}word2vec_embeddings.vec', 'rb') as f:
    embeddings_dim = int(next(f).split()[1])
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype=np.float32)
        embeddings[word] = coefs

with open(f'{DATA_DIR}word_index.pickle', 'rb') as f:
    word_index = pickle.load(f)

print('Vocabulary:', len(word_index))

# Compute embedding matrix
embedding_matrix = np.zeros((len(word_index) + 1, embeddings_dim))
unknown_count = 0
for idx, (word, i) in enumerate(word_index.items()):
    vec = embeddings.get(word)
    if vec is not None:
        embedding_matrix[i] = vec
    else:
        print(word)
        unknown_count += 1

print('Unknown:', unknown_count)

np.save(f'{DATA_DIR}embedding_matrix.npy', embedding_matrix)
```
